```python
data.drop(['rv1', 'rv2'], axis=1, inplace=True)
```

```python
data.head()
```

```python
data_date = data['date']
```

```python
date_value = data_date.values
```

```python
y_m_d = []
h_m_s = []
for index in date_value:
    str1 = index.split()
    y_m_d.append(str1[0])
    h_m_s.append(str1[1])

NSM = []
# 计算NSM 这是这一天的第多少秒
for index in h_m_s:
    str1 = index.split(':')

    NSM.append(int(str1[0]) * 3600 + int(str1[1]) * 60 + int(str1[2]) * 1)


data['NSM'] = NSM

print(data.head())
```

# 标注，这一天是周几

```python
month = [31, 29, 31, 30, 31, 30, 31, 31, 30, 31,30,  31]
sum(month)
# for index in y_m_d:
#     str1 = index.split('-')
    
```

如何根据2016-01-11是周一推算以后每一天是周几

```python
# 规定2016-1-1为周五用5表示，
flag = 5
week_index = []
sate_of_week = []
for index in y_m_d:
    str1 = index.split('-')
    count_day = int(str1[2])
    for i in range(int(str1[1])-1):
        count_day += month[i]
    temp_point = (count_day % 7) + 4

    if temp_point <= 7:
        week_flag = temp_point
    else:
        week_flag = temp_point - 7
    week_index.append(week_flag)
    
    if week_flag > 5:
        sate_of_week.append('weekned')
    else:
        sate_of_week.append('work')

data['day_of_week'] = week_index
data['sate_of_week'] = sate_of_week
data['day_of_week'].replace([1, 2, 3, 4, 5, 6, 7],['Monday', 'Tuesday', 'Wednesday ', 'Thursday', 'Friday', 'Saturday', 'Sunday'],inplace=True)
```

数据的编码

```python
a = ['day_of_week', 'sate_of_week']
data_yabianam = pd.get_dummies(data[a])
data_yabianam
```

```python
data.drop(['date', 'day_of_week', 'sate_of_week'], axis=1, inplace=True)
```

```python
data_result = pd.concat([data, data_yabianam], axis=1)
data_result.head()
```

```python
data_result.to_csv('data_worked_num.csv',index=False)
```
